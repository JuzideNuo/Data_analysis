```python
# 干掉索引 那些奇怪的东西
fz_demo = ['city', 'fields.comment', 'fields.discountPrice', 'fields.features', 'fields.itemId',
           'fields.itemTotalScore', 'fields.latitude', 'fields.longitude', 'fields.price',
           'fields.shortInfo', 'fields.sold365', 'fields.soldRecentNum', 'fields.tagList', 
            'fields.title', 'trip_main_busness_type']
fz = fz[fz_demo]
fz.head()
```

```python
fz = fz.fillna(0)
fz.head()
```

```python
fz.drop_duplicates(subset='fields.itemId', inplace=True)
fz.shape
```

```python
fz.dtypes
```

```python
fz['fields.features'] = fz['fields.features'].replace('"','',regex=True)
```

```python
# fields.sold365清洗
def get_sold365(fz):
    if fz is None:
        return np.nan
    if "万" in fz:
        return float(fz[fz.find('售')+1:fz.find('万')]) * 10000
    else:
        return float(fz[fz.find('售')+1:fz.find('笔')])

    
fz['fields.sold365'] = fz['fields.sold365'].apply(get_sold365)
fz.head()
```
