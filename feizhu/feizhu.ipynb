```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient
from pandas.io.json import json_normalize

plt.style.use('ggplot')
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']  #解决seaborn中文字体显示问题
plt.rc('figure', figsize=(10, 10))  #把plt默认的图片size调大一点
plt.rcParams["figure.dpi"] =mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
%matplotlib inline
```

```python
conn = MongoClient(host='127.0.0.1', port=27017)  # 实例化MongoClient
db = conn.get_database('Laborday')  

col = db.get_collection('ticket') # 连接到集合ticket
mon_data = col.find()  # 查询这个集合下的所有记录
```

```python
fz = json_normalize([rec for rec in mon_data])
display(fz.shape,fz.head())
```

```python
fz_keep = ['city', 'fields.comment', 'fields.discountPrice', 'fields.features', 'fields.itemId',
           'fields.itemTotalScore', 'fields.latitude', 'fields.longitude', 'fields.price',
           'fields.shortInfo', 'fields.sold365', 'fields.soldRecentNum', 'fields.tagList', 
            'fields.title', 'trip_main_busness_type']
fz = fz[fz_keep]
fz
```

```python
fz = fz.fillna(0)
fz
```

```python
fz.drop_duplicates(subset='fields.itemId', inplace=True)
fz.shape
```

```python
fz.dtypes
```

```python
fz.head()
```

```python
fz['fields.comment'] = fz['fields.comment'].astype(np.float64)
```

```python
fz['fields.discountPrice'] = fz['fields.discountPrice'].astype(np.float64)
```

```python
fz['fields.itemId'] = fz['fields.itemId'].astype(np.int64)
```

```python
fz['fields.price'] = fz['fields.price'].astype(np.float64)
fz['fields.soldRecentNum'] = fz['fields.soldRecentNum'].astype(np.int)
fz.dtypes
```

```python
fz.head()
```

```python
new_fz = fz
new_fz.head()
```

```python
def get_fea(data):
    if len(data) > 0:
        return data[0]['text']
    else:
        return None

new_fz['fields.features'] = new_fz['fields.features'].apply(get_fea)
new_fz.head()
```

```python
def get_sold365(new_fz):
    if new_fz != 0:
        if "万" in new_fz:
            return float(new_fz[new_fz.find('售')+1:new_fz.find('万')]) * 10000
        else:
            return float(new_fz[new_fz.find('售')+1:new_fz.find('笔')])

    
new_fz['fields.sold365'] = new_fz['fields.sold365'].apply(get_sold365)
new_fz.sample(5)
```

```python
new_fz = new_fz.fillna(0)
new_fz
```

```python
new_fz['fields.tagList'].sample(5)
```

```python
new_fz['fields.tagList'] = new_fz['fields.tagList'].apply(get_fea)
new_fz['fields.tagList'].sample(5)
```

```python
new_fz = new_fz.fillna(0)
new_fz['fields.tagList'].sample(5)
```

```python
city = new_fz.merge(city_data, on='city')
city.head()
```
